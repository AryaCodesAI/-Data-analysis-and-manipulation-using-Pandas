### Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Assign it to a variable called users.

```python
import pandas as pd
df_users= pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
df_users
```

### Discover what is the mean age per occupation

```python
df_users.groupby('occupation').age.mean()
```

### Discover the Male ratio per occupation and sort it from the most to the least

```python
df_users.groupby(['occupation', 'gender']).age.sum()
```

```python
df_users.groupby('occupation').gender.mean()
```

### For each occupation, calculate the minimum and maximum ages

```python
data= {min : df_users.groupby('occupation').age.min(),
max : df_users.groupby('occupation').age.max()}
df_minmax = pd.DataFrame(data)
df_minmax
```

### For each combination of occupation and gender, calculate the mean age

```python
df_users.groupby(['occupation', 'gender']).age.mean()
```

###  For each occupation present the percentage of women and men

```python
df_users.groupby(['occupation', 'gender']).max()
```
